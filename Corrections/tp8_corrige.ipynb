## Exercice 1 : Résumé des météos de Paris en 2009

Reprendre les données du TP6. Pour chaque mois, créer un dictionnaire contenant les entrées suivantes :

- `nom` : nom du mois,
- `T` : moyenne des températures minimales et maximales,
- `précipitations` : moyenne des précipitations.

Créer un tableau de douze dictionnaires contenant les informations des différents mois.

```python
##################
#   Correction   #
##################


def moyenne_mois(meteoMois):
    temperature = 0.
    precipitations = 0.
    i = 0
    while i < len(meteoMois):
        temperature += (meteoMois[i]["Tmin"] + meteoMois[i]["Tmax"]) / 2.
        precipitations += meteoMois[i]["précipitation"]
        i += 1
    return [ temperature / len(meteoMois), precipitations / len(meteoMois)]
```

```python
##################
#   Correction   #
##################


import json

f = open("data/meteo_paris_2009.json")
str = f.read()
f.close()

meteo2009 = json.loads(str)



resume = []

mois = list(meteo2009)
i = 0
while i < len(mois):
    m = moyenne_mois(meteo2009[mois[i]])
    resume.append({
        "nom": mois[i],
        "T" : m[0],
        "precipitations":m[1]
    })
    i += 1


print(resume)
```

## Exercice 2 : Comparaison de mois

Définir une fonction `compare_temperatures` prenant en paramètre deux dictionnaires (comme ceux de l'exercice 1) et retournant :

- -1 si la température du premier dictionnaire est inférieure strictement à celle du deuxième, 
- 0 si les deux températures sont égales,
- 1 si la température du premier dictionnaire est supérieure strictement à celle du deuxième.


```python
##################
#   Correction   #
##################


import math

def compare_temperatures(m1, m2):
    if math.isclose(m1["T"], m2["T"]):
        return 0
    elif m1["T"] < m2["T"]:
        return -1
    else:
        return 1
```

## Exercice 3 : Tri selon la température

- Définir la fonction `tri_temperature` permettant de trier un tableau de dictionnaires (comme ceux de l'exercice 1) et triant le tableau dans l'ordre croissant selon la température des mois.
- Afficher les mois de l'année 2009 du plus froid au plus chaud (en moyenne).

```python
##################
#   Correction   #
##################


def tri_temperature(tab):
    """
        Trie le tableau passé en paramètre selon la température à l'aide du tri par sélection.
    """
    
    # définit la position au-delà de laquelle le tableau n'est
    # pas encore trié
    pos_debut_non_trie = 0
    while pos_debut_non_trie < len(tab) - 1:
        # recherche de la position du minimum à partir de pos_debut_non_trie
        i = pos_debut_non_trie
        pos_min = i
        while i < len(tab):
            if compare_temperatures(tab[i], tab[pos_min]) == -1:
                pos_min = i
            i += 1
        # on place la valeur min à sa place à la fin de la partie déjà triée
        # en l'intervertissant avec la valeur de la première case non triée
        tmp = tab[pos_min]
        tab[pos_min] = tab[pos_debut_non_trie]
        tab[pos_debut_non_trie] = tmp
        # on décale la position de debut de tableau non trié
        pos_debut_non_trie += 1
        
tri_temperature(resume)
i = 0
while i < len(resume):
    print("-", resume[i]["nom"])
    i += 1
```

## Exercice 4 : Comparaison selon les précipitations

```python
##################
#   Correction   #
##################


import math

def compare_precipitations(m1, m2):
    if math.isclose(m1["precipitations"], m2["precipitations"]):
        return 0
    elif m1["precipitations"] < m2["precipitations"]:
        return -1
    else:
        return 1
```
