## Exercice 1 : Opérations de base dans une 1ère table de hachage 

### Question 1  :  La recherche

- Définir une fonction `rechercher` qui prend en paramètre une table de hachage et un entier et renvoie `True` si l'entier est dans la table, et `False` sinon. 

```python
##################
#   Correction   #
##################


def rechercher(table, entier):
    if entier >= 0:
        indice = entier % 10
    else:
        indice = -entier % 10
    j = 0
    while j < len(table[indice]):
        if table[indice][j] == entier:
            return True
        j += 1
    return False
```

### Question 2 : L'ajout  

- Définir une fonction `ajouter` qui prend en paramètre une table de hachage et un entier et ajoute cet entier dans la table s'il n'existe pas. 

```python
##################
#   Correction   #
##################


def ajouter(table, entier):
    if not rechercher(table, entier):
        if entier >= 0:
            indice = entier % 10
        else:
            indice = -entier % 10
        table[indice].append(entier)
```

### Question  3  : La suppression

- Définir une fonction `supprimer` qui prend en paramètre une table de hachage et un entier et supprime cet entier dans la table s'il y est présent.

**Remarque :** On utilisera la fonction `L.pop()` qui supprime le dernier élément de la table `L`.

```python
##################
#   Correction   #
##################


def supprimer(table, entier):
    if rechercher(table, entier):
        if entier >= 0:
            indice = entier % 10
        else:
            indice = -entier % 10

        # On cherche la position de l'entier dans table[indice]
        j = 0
        while j < len(table[indice]) and table[indice][j] != entier:
            j += 1
        #On met le dernier élément de table[indice] à la place de entier
        table[indice][j] = table[indice][len(table[indice]) - 1]
        table[indice].pop()  # On supprime le dernier élément de table[indice]
```

**Remarque :** Si la valeur que l'on veut stocker dans la table est entière, il suffit de transformer l'entier en une chaîne de caractères avec `str(entier)`.

```python
# Fonction de hachage définie en cours

def hachage(str):
    nombre = 0
    i = 0
    while i < len(str):
        nombre += ord(str[i])
        i += 1
    return nombre
```

## Exercice 3 : Hachage fermé (ou à adressage ouvert)

On va dans cet exercice étudier une méthode pour gérer les *collisions* (cas où plusieurs éléments à stocker ont la même valeur de hachage) en s'autorisant à stocker au plus un élément par case et non plus une liste arbitrairement longue d'éléments.

Lorque l'on essaie de d'insérer un nouvel est élément et que la case où on veut le mettre est déjà occupée, il faut essayer de le placer dans un autre case, et ainsi de suite jusqu'à en trouver une libre. On dit alors que l'on *sonde* la table pour trouver une case libre. Dans ces conditions, la fonction de hachage prend un argument supplémentaire qui compte le nombre de tentatives infructueuses pour trouver une case libre.

### Question 1 : Sondage linéaire 

Dans le cas du *sondage linéaire*, on avance dans le tableau d'un nombre fixe de cases (souvent une case) à chaque fois jusqu'à trouver une case vide. Si le pas est un, la fonction de hachage s'écrit $h(k,i)= h_1(k)+i$ où $h_1$ est un fonction de hachage simple et $i$ est le nombre de tentatives pour trouver une case vide.

- Éxécuter à la main l'algorithme d'insertion dans une table de taille `m=9` des  valeurs 5, 28, 19, 15, 20, 33, 12, 17 en utilsant la fonction $h_1(k)=k \mod 9$.
- Écrire les fonctions permettant d'insérer et de rechercher une valeur dans une table de hachage de taille `m` avec la fonction de hachage à sondage linéaire qui vient d'être décrite. 
- Que se passe-t-il si l'on supprime des valeurs dans la table ? Comment peut-on résoudre ce problème ?

```python
##################
#   Correction   #
##################


#la table est initialisée des cases contenant None
def rechercher3(table, entier):
    indice = entier % len(table)
    i = indice
    j = 0
    while table[i] != entier and table[i] != None and j != len(table):
        i = (i + 1) % len(table)
        j += 1
    if table[i] == entier:
        return True
    return False
```

```python
##################
#   Correction   #
##################


def ajouter3(table, valeur):
    if not rechercher3(table, valeur):
        indice = valeur % len(table)
        while table[indice] != None:
            indice += 1
        table[indice] = valeur
```

**CORRECTION :**


Si l'on supprime des valeurs, il ne faut pas pour autant arrêter la recherche quand on rencontre une case qui avait été occupée. Le sondage peut avoir conduit à stocker la valeur au-delà de cette case.

Si la suppression est implantée, il faut donc que l'on puisse savoir qu'une case a déjà été occupée. Si c'est le cas, la recherche doit continuer au delà de cette case.

### Question 2 : Double hachage

Le sondage linéaire présente un gros inconvénient. Même si la fonction de hachage répartit les clés uniformément dans la table, la résolution des collisions a tendance à former des "paquets" de valeurs ce qui conduit à rapidement augmenter le nombre de recherches infructueuses.

Une solution pour résoudre le problème des "paquets" est de modifier la méthode de sondage pour que les cases sondées soient éloignées de manière variable. Le double hachage est l'une des meilleurs méthodes connues pour cela. Il utilise une fonction de hachage de la forme $h(k,i)=h_1(k)+ih_2(k) \mod m$ où $h_1$  et $h_2$ sont des  fonctions de hachage auxiliaires simples.

Si $h_2(k)$ et $m$ ne sont pas premiers entre eux, on ne parcourt pas toutes les cases possibles dans le tableau.
Par exemple si les deux valeurs ont comme PGCD 2, on parcourt que la moitié des cases du tableau. D'une manière générale on ne parcourt qu'une proportion 1/PGCD des cases du tableau. On peut donc échouer à trouver une case vide alors que le tableau n'est pas du tout plein.
