## Exercice 1 : le jeu du devin

On propose de programmer le jeu du devin. Les règles sont simples :
vous choisissez (mentalement) un nombre entre 1 et 100 et le devin (c'est-à-dire l'ordinateur) doit le trouver
en un minimum d'essais. À chaque essai, l'ordinateur propose un nombre et
vous devez répondre (sans tricher !) s'il a trouvé (le programme
est alors terminé), ou si votre nombre est plus grand ou plus petit que le nombre proposé.

```python
##################
#   Correction   #
##################


def saisie():
    print("Veuillez saisir votre réponse : ")
    print("1 - Bravo devin, tu as trouvé mon nombre !")
    print("2 - Non, mon nombre est plus grand")
    print("3 - Nom, mon nombre est plus petit")
    n = -1
    while n < 1 or n > 3:
        n = int(input())
    return n


def jeu_devin():
    debut = 1
    fin = 100
    milieu = (debut + fin) // 2
    trouve = False
    while not trouve and debut <= fin:
        print("Est-ce qu le nombre choisi est ", milieu, "?")
        res = saisie()
        if res == 1:  #On a trouvé l'élément
            return
        elif res == 3:
            fin = milieu - 1
        else:
            debut = milieu + 1
        milieu = (debut + fin) // 2
    if debut > fin:
        print("Certains trichent !!!")


jeu_devin()
```

## Exercice 2 : Mêmes caractères

Définir la fonction `memes_caracteres` prenant deux chaînes
et renvoyant `True` si elles comportent les mêmes caractères
(tout caractère de l'une apparaît dans l'autre), et `False` sinon.

```python
##################
#   Correction   #
##################


def appartient(caractere, ch):
    i = 0
    rep = False
    while i < len(ch) and not rep:
        if ch[i] == caractere:
            rep = True
        i += 1
    return rep


def inclus(ch1, ch2):
    i = 0
    n = len(ch1)
    while i < n:
        if not appartient(ch1[i], ch2):
            return False
        i += 1
    return True


def memes_caracteres(ch1, ch2):
    return inclus(ch1, ch2) and inclus(ch2, ch1)


c1 = "la crise économique"
c2 = "le scénario comique "

if memes_caracteres(c1, c2):
    print(c1, "et", c2, "utilisent les même caractères")
else:
    print("Tous les caractères n'apparaissent pas dans", c1, " et", c2)
```

Définir la fonction `memes_caracteres_trie` prenant deux chaînes
**triées** et renvoyant `True` si elles comportent les mêmes
caractères et `False` sinon.

```python
##################
#   Correction   #
##################


def memes_caracteres_trie(ch1, ch2):
    if len(ch1) == 0 and len(ch2) == 0:
        return True
    elif len(ch1) == 0 or len(ch2) == 0:
        return False

    i = 0
    j = 0
    while i < len(ch1) and j < len(ch2):
        if ch1[i] != ch2[j]:
            return False
        i += 1  # on passe au prochain (nouveau) caractère dans ch1
        while i < len(ch1) and ch1[i] == ch1[i - 1]:
            i += 1
        j += 1
        while j < len(ch2) and ch2[j] == ch2[j - 1]:
            j += 1
    return i == len(ch1) and j == len(ch2)


c1 = "aabbbbbbbbbbbbbbbbbbc"
c2 = "aabbccc"

if memes_caracteres_trie(c1, c2):
    print(c1, "et", c2, "utilisent les même caractères")
else:
    print("Tous les caractères n'apparaissent pas dans", c1, " et", c2)
```

Quelle est la complexité dans le pire des cas de chacune des fonctions ?

**CORRECTION :**


- non trié : $O(nm)$ avec $n$ et $m$ les tailles des chaînes
- trié : $O(n)$ avec $n$ la taille de la plus longue des deux chaînes

## Exercice 3 : Calcul d'une racine carrée

Soit une fonction mathématique $f$ continue et strictement
monotone (comme la racine carrée). Si $f(a)$ et $f(b)$ sont de
signe différent (soit $f(a) \times f(b) ≤ 0$), le théorème des
valeurs intermédiaires garantit l’existence d’un nombre $x$ tel
que $f(x) = 0$ sur l’intervalle $[a,b]$.
On peut rechercher ce nombre par dichotomie, de manière approchée :
    
- on calcule $c = (a+b)/2$ et $f(c)$.
- si $f(a) \times f(c) ≤ 0$, la solution $x$ appartient à
l’intervalle $[a,c]$ ; sinon elle appartient à l’intervalle $[c,b]$.
- dans le premier cas, on remplace $b$ par $c$ ; dans le second,
on remplace $a$ par $c$.

On répète ces étapes tant que la longeur de l’intervalle est
supérieure à une précision $\epsilon$ donnée.
La valeur de $c$ approchera alors la solution de $f(x)=0$ à
$\epsilon/2$ près.

Définir une fonction calculant la racine carrée d'un nombre.

```python
##################
#   Correction   #
##################


def f(x, nb):
    return x * x - nb


def racine(nb):
    if nb < 0:
        return -1
    deb = 0
    
    # si nb < 1 alors sa racine carrée est comprise entre nb et 1
    # sinon, elle est inférieure à nb
    if nb >= 1:
        fin = nb
    else:
        fin = 1
    
    while fin - deb > 0.00001:
        milieu = (fin + deb) / 2
        if f(deb, nb) * f(milieu, nb) <= 0:
            fin = milieu
        else:
            deb = milieu
    return (fin + deb) / 2


nombre = 110
res = racine(nombre)
print("Racine de", nombre, "=", res)
```
